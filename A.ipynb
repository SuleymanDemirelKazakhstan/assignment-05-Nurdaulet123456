```python
import pandas as pd
import numpy as np
from numpy import loadtxt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical, plot_model

df = pd.read_csv('./test.csv', on_bad_lines='skip')
```

```python
X = df.iloc[:, 1:]
Y = to_categorical(df.iloc[:, 0])

print(Y)
```

```python
modal = Sequential()
modal.add(Dense(12, input_shape=(784,), activation='relu'))
modal.add(Dense(8, activation='relu'))
modal.add(Dense(10, activation='sigmoid'))
```

```python
modal.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
modal.fit(X, Y, epochs=150, batch_size=10)
```
